## Dividend Chart Check

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV  , index_col=None 
dividend = pd.read_csv('final_data/final_data.ipynb.csv', index_col=None)

# 10x7 subplot 
fig, axs = plt.subplots(10, 7, figsize=(20, 30))

# dividend     subplot  
for i in range(66):
    row = i // 7  #   
    col = i % 7   #   
    
    sample = dividend.iloc[:, i]
    ssample = sample[sample != 0]
    
    axs[row, col].scatter(ssample.index, ssample)
    axs[row, col].set_xlabel('Date')
    axs[row, col].set_ylabel(f'{ssample.name}_dividend')
    axs[row, col].grid(True)

plt.tight_layout()
plt.show()
```
