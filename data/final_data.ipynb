### BEN, BF-B, PNR, TROW 수정

```python
datacopy = data.copy()
```

```python
BEN = pd.read_csv('BEN_d_revised.csv')

BEN['Date'] = pd.to_datetime(BEN['Date'])

BEN['Date'] = BEN['Date'].dt.strftime('%Y-%m')

BEN = BEN[(BEN['Date'] >= '2014-02') & (BEN['Date'] <= '2024-02')]


BEN
```

```python
BFB = pd.read_csv('BF-B_d_revised.csv')

BFB['Date'] = pd.to_datetime(BFB['Date'])

BFB['Date'] = BFB['Date'].dt.strftime('%Y-%m')

BFB = BFB[(BFB['Date'] >= '2014-02') & (BFB['Date'] <= '2024-02')]


BFB
```

```python
PNR = pd.read_csv('PNR_d_revised.csv')

PNR['Date'] = pd.to_datetime(PNR['Date'])

PNR['Date'] = PNR['Date'].dt.strftime('%Y-%m')

PNR = PNR[(PNR['Date'] >= '2014-02') & (PNR['Date'] <= '2024-02')]


PNR
```

```python
TROW = pd.read_csv('TROW_d_revised.csv')

TROW['Date'] = pd.to_datetime(TROW['Date'])

TROW['Date'] = TROW['Date'].dt.strftime('%Y-%m')

TROW = TROW[(TROW['Date'] >= '2014-02') & (TROW['Date'] <= '2024-02')]


TROW
```

```python
data = data.drop(columns = ['BEN', 'BF-B', 'PNR', 'TROW'])
```

```python
data.columns
```

```python
merged_data = data.copy()

pd.merge(BEN, BFB, how = 'left', on = 'Date')
```

```python


def prepare_dividend_df(df, ticker_name):
    df = df.rename(columns={'Dividends': ticker_name})
    df.set_index('Date', inplace=True)
    return df[[ticker_name]]

ben_df = prepare_dividend_df(BEN, 'BEN')
bfb_df = prepare_dividend_df(BFB, 'BFB')
pnr_df = prepare_dividend_df(PNR, 'PNR')
trow_df = prepare_dividend_df(TROW, 'TROW')


data.set_index('Date', inplace=True)


merged_data = data.copy()
merged_data = merged_data.join(ben_df, how='left')
merged_data = merged_data.join(bfb_df, how='left')
merged_data = merged_data.join(pnr_df, how='left')
merged_data = merged_data.join(trow_df, how='left')


merged_data = merged_data.reindex(sorted(merged_data.columns), axis=1)


merged_data
```

```python
merged_data = merged_data.fillna(0)
```

```python
final_data = merged_data


final_data
```

```python
output_path = 'final_data.csv'
final_data.to_csv(output_path, index=True)
```
